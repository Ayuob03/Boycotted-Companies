```python
import tkinter as tk
from tkinter import filedialog, messagebox, simpledialog
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool

# Function to load dataset
def load_dataset():
    global df
    file_path = filedialog.askopenfilename(title="Select a CSV File", filetypes=[("CSV Files", "*.csv")])
    if file_path:
        try:
            df = pd.read_csv(file_path)
            df.columns = df.columns.str.strip().str.lower()
            if 'date' in df.columns:
                df['date'] = pd.to_datetime(df['date'], errors='coerce')
            messagebox.showinfo("Success", "File loaded successfully!")
            print("Columns in the dataset:", df.columns)
        except Exception as e:
            messagebox.showerror("Error", f"Failed to load file: {e}")
    else:
        messagebox.showwarning("Warning", "No file selected!")

# Function to display basic statistics
def display_statistics():
    global df
    if df is None:
        messagebox.showwarning("Warning", "Please load a dataset first!")
        return
    try:
        # إنشاء نافذة جديدة لعرض الإحصاءات
        stats_window = tk.Toplevel()
        stats_window.title("Dataset Statistics")
        stats_window.geometry("600x400")  # تعيين حجم النافذة
        stats_window.configure(bg="#f5f5f5")  # خلفية بلون فاتح

        # إضافة عنوان للنافذة
        title_label = tk.Label(stats_window, text="Dataset Statistics", font=("Arial", 16, "bold"), bg="#f5f5f5", fg="#333333")
        title_label.pack(pady=10)

        # إنشاء صندوق نصوص لعرض الإحصاءات
        text_area = tk.Text(stats_window, wrap="word", font=("Consolas", 12), bg="#ffffff", fg="#333333", padx=10, pady=10)
        text_area.pack(expand=True, fill="both", padx=10, pady=10)

        # إضافة شريط تمرير
        scroll = tk.Scrollbar(text_area, command=text_area.yview)
        text_area.configure(yscrollcommand=scroll.set)
        scroll.pack(side="right", fill="y")

        # عرض الإحصاءات داخل صندوق النصوص
        stats = df.describe(include='all').to_string()
        text_area.insert("1.0", stats)
        text_area.configure(state="disabled")  # جعل النص غير قابل للتعديل

        # زر إغلاق النافذة
        close_button = tk.Button(stats_window, text="Close", command=stats_window.destroy, bg="#0078D7", fg="white", font=("Arial", 12), relief="raised", padx=10, pady=5)
        close_button.pack(pady=10)

    except Exception as e:
        messagebox.showerror("Error", f"Failed to calculate statistics: {e}")

# Function to visualize data using an interactive LineChart with Bokeh
def visualize_data():
    global df
    if df is None:
        messagebox.showwarning("Warning", "Please load a dataset first!")
        return
    try:
        if 'date' not in df.columns or 'avg price' not in df.columns:
            messagebox.showerror("Error", "The dataset must contain 'date' and 'avg price' columns.")
            return

        output_file("interactive_line_chart.html")

        p = figure(
            title="Stock AVG Price Over Time",
            x_axis_label='Date',
            y_axis_label='AVG Price',
            x_axis_type='datetime',
            width=900,
            height=500,
            tools="pan,box_zoom,reset,save"
        )

        p.line(df['date'], df['avg price'], line_width=2, color="blue", legend_label="AVG Price")
        p.circle(df['date'], df['avg price'], size=6, color="blue")

        hover = HoverTool(
            tooltips=[
                ("Date", "@x{%F}"),
                ("AVG Price", "$@y{0.00}")
            ],
            formatters={"@x": "datetime"},
            mode='vline'
        )
        p.add_tools(hover)

        show(p)

    except Exception as e:
        messagebox.showerror("Error", f"Failed to visualize data: {e}")

# Function to process data
def process_data():
    global df
    if df is None:
        messagebox.showwarning("Warning", "Please load a dataset first!")
        return
    try:
        df['year'] = df['date'].dt.year
        df['month'] = df['date'].dt.month
        df['day'] = df['date'].dt.day
        messagebox.showinfo("Success", "Data processed successfully!")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to process data: {e}")

# Function to filter data by month
def filter_by_month():
    global df
    if df is None:
        messagebox.showwarning("Warning", "Please load a dataset first!")
        return
    try:
        month = simpledialog.askinteger("Filter by Month", "Enter a month number (1-12):")
        if month not in range(1, 13):
            messagebox.showerror("Error", "Invalid month. Please enter a number between 1 and 12.")
            return
        filtered_df = df[df['month'] == month]
        if filtered_df.empty:
            messagebox.showinfo("No Data", f"No data available for month {month}.")
        else:
            messagebox.showinfo("Filter Success", f"Filtered data for month {month}.")
            print(filtered_df.head())
    except Exception as e:
        messagebox.showerror("Error", f"Failed to filter data: {e}")

# Function to visualize trading volume
def visualize_volume():
    global df
    if df is None:
        messagebox.showwarning("Warning", "Please load a dataset first!")
        return
    try:
        if 'date' not in df.columns or 'volume' not in df.columns:
            messagebox.showerror("Error", "The dataset must contain 'date' and 'volume' columns.")
            return

        output_file("volume_chart.html")
        p = figure(
            title="Trading Volume Over Time",
            x_axis_label='Date',
            y_axis_label='Volume',
            x_axis_type='datetime',
            width=900,
            height=500,
            tools="pan,box_zoom,reset,save"
        )
        p.vbar(x=df['date'], top=df['volume'], width=0.9, color="blue", legend_label="Volume")

        hover = HoverTool(
            tooltips=[("Date", "@x{%F}"), ("Volume", "@top")],
            formatters={"@x": "datetime"},
            mode='vline'
        )
        p.add_tools(hover)

        p.legend.title = "Data"
        show(p)

    except Exception as e:
        messagebox.showerror("Error", f"Failed to visualize volume: {e}")

# Function to change button color on hover
def on_hover(button, color_on_hover, color_on_leave):
    button.bind("<Enter>", lambda event: button.config(bg=color_on_hover))
    button.bind("<Leave>", lambda event: button.config(bg=color_on_leave))

# Initialize the main application window
root = tk.Tk()
root.title("Stock Analysis GUI")
root.geometry("500x600")
root.configure(bg='#e6f2ff')  # Light blue background

df = None  # Global dataframe to store the loaded data

# Styling the buttons
button_style = {"padx": 10, "pady": 5, "bd": 2, "bg": "#0078D7", "fg": "white", "font": ("Arial", 12), "relief": "raised"}

# Title Label
title_label = tk.Label(root, text="Stock Price Analysis Tool", bg='#e6f2ff', fg='#333333', font=("Arial", 18, "bold"), relief="groove", padx=10, pady=5)
title_label.pack(pady=15)

# Add buttons to the interface
btn_load = tk.Button(root, text="Load Dataset", command=load_dataset, **button_style)
btn_process = tk.Button(root, text="Process Data", command=process_data, **button_style)
btn_visualize = tk.Button(root, text="Visualize Data", command=visualize_data, **button_style)
btn_filter = tk.Button(root, text="Filter by Month", command=filter_by_month, **button_style)
btn_stats = tk.Button(root, text="Display Statistics", command=display_statistics, **button_style)
btn_volume = tk.Button(root, text="Visualize Volume", command=visualize_volume, **button_style)
btn_exit = tk.Button(root, text="Exit", command=root.quit, **button_style)

# Adding hover effect to buttons
on_hover(btn_load, "#005BB5", "#0078D7")
on_hover(btn_process, "#005BB5", "#0078D7")
on_hover(btn_visualize, "#005BB5", "#0078D7")
on_hover(btn_filter, "#005BB5", "#0078D7")
on_hover(btn_stats, "#005BB5", "#0078D7")
on_hover(btn_volume, "#005BB5", "#0078D7")
on_hover(btn_exit, "#990000", "#0078D7")

# Place buttons on the GUI
btn_load.pack(pady=10)
btn_process.pack(pady=10)
btn_visualize.pack(pady=10)
btn_filter.pack(pady=10)
btn_stats.pack(pady=10)
btn_volume.pack(pady=10)
btn_exit.pack(pady=10)

# Run the main event loop
root.mainloop()
```
